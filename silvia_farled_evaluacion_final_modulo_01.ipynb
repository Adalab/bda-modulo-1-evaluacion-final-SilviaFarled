Crear una clase llamada TiendaOnline

 1. `inventario` (lista de diccionarios): Un atributo para almacenar los productos en el inventario. 
  Cada producto debe ser representado como un diccionario con las siguientes claves: `'nombre'`, `'precio'`, y `'cantidad'`.
   Al principio deberá ser una lista vacía. Ejemplo de como debería ser:
      ```python
      [{'nombre': 'Camisa', 'precio': 20, 'cantidad': 40},
      {'nombre': 'Pantalón', 'precio': 30, 'cantidad': 30}]
      ```
  2. `clientes` (diccionario): Un atributo para llevar un registro de los clientes de la tienda. 
  Cada cliente debe ser representado como un diccionario con las siguientes claves: `'nombre'` y `'email'`. 
  Al inicio deberá ser un diccionario vacío. Además, cada cliente debe tener un historial de compras. Deberá parecerse a:
      ```python
      {'Cliente1': {'email': 'cliente1@email.com', 'compras': []},
      'Cliente2': {'email': 'cliente2@email.com', 'compras': []}}
      ```
  3. `ventas_totales` (float): Un atributo para llevar un registro de las ventas totales de la tienda. 
  Inicializalo con valor 0. 

````python
# Crear una clase llamada TiendaOnline con 3 atributos: inventario (diccionario), clientes (diccionario) y ventas_totales:

class TiendaOnline:

    def __init__(self, inventory, customer, total_sales=0.0):
        self.inventory = inventory
        self.customer = customer
        self.total_sales = total_sales

# Al ser atributos vacíos, los definimos como lista, diccionario o float
#    def init(self): self.inventory = []
#    def init(self): self.customer = {}



# 1. Definimos una primera función para actualizar el inventario con un nuevo producto o bien con sumando la cantidad de uno ya exitente
    def add_product(self, name, price, stock):
        for product in self.inventory:
            if product["name"] == name:
               product["stock"] += stock
               return f"Se ha actualizado el stock de {name} correctamente."
        else:
            self.inventory.append({"name":name, "price":price, "stock": stock})
            return f"Se ha añadido el producto {name} correctamente al inventario"

# 2. Mostramos el inventario de productos con sus detalles. 

    def check_inventory(self):
        inventory_list = []
        for product in self.inventory:
            inventory_list.append(f"Nombre: {product['name']}, Precio: {product['price']}, Cantidad:{product['stock']}")
        return "\n".join(inventory_list)
            

# 3. Buscamos un producto indicando el nombre y recibimos toda su información
    def search_product(self, name):
        for product in self.inventory: 
            if product["name"] == name:
                print(f"Nombre: {product['name']}, Precio: {product['price']}, Cantidad:{product['stock']}")
                break
        else:
            print(f"El producto {name} no está en el inventario")


# 4. Actualiza el stock de un producto en el inventario. 
    def update_stock(self,name,stock):
        for product in self.inventory:
            if product["name"] == name:
                product["stock"] += stock
                return f"El stock de {name} ha sido actualizado correctamente. Actualmente hay {product['stock']}"
        else: 
            print(f"No es posible actualizar el producto {name} porque no está en el inventario")

        
# 5. Elimina un producto del inventario. 
    def delete_product(self, name): 
        for product in self.inventory: 
            if product["name"] == name:
                self.inventory.remove(product)
                return f"El producto {name} ha sido eliminado del inventario correctamente."
        else: 
            print(f"El producto {name} no se puede eliminar porque no está en el inventario")

# 6. Calcula el valor total de los productos del inventario. 
    def inventory_value(self): 
        total_value = 0
        for product in self.inventory: 
            value = product["price"]*product["stock"]
            total_value += value
        return total_value

# 7. Realizar una compra con carrito, bule while, mostrar inventario, opción de salir, registrar los productos en el carrito y calcular el coste de la compra. 
    def purchase(self):
        cart = {}
        print("Bienvenida a nuestra tienda.")
        print(f"Estos son los productos disponibles:\n{self.check_inventory()}.")
        order = (input("Por favor, introduzca el producto que desea adquirir:\n Si no desea continuar comprando escriba la palabra 'Salir'")).lower()                      
        
        try: 
            while order != "salir":
                for product in self.inventory: 
                    if (product["name"]).lower() == order:
                        number = int(input(f"¿Qué cantidad de {order} desea comprar?"))
                        if product["stock"] >= number and number > 0:
                            cart[order]=cart.get(order,0) + number
                            product["stock"] -= number
                            print(f"Producto agregado al carrito. {product['stock']} unidades de {order} restantes.")
                        else:
                            print(f"No hay suficiente stock de {order}. Solo hay {product['stock']} unidades disponibles.")
                        break
                else: 
                    print("El producto seleccionado no está disponible en nuestra tienda")
                order = (input("Por favor, introduzca el producto que desea adquirir:\n Si no desea continuar comprando escriba la palabra 'Salir'")).lower()
            if order == "salir":     
                if len(cart) == 0:
                    print("Su compra ha finalizado y su carrito está vacío.")
                else:
                    total_cost = sum(product["price"] * cart[order] for order in cart for product in self.inventory if product["name"].lower() == order)
                print(f"Ha comprado los siguientes productos: {cart}")
                print(f"El costo total de su compra es: {total_cost:.2f}")
                print("Muchas gracias por elegirnos, esperamos que vuelva de nuevo para probar otro instrumento")
        except ValueError:
            print("Error: por favor, ingrese un número válido para la cantidad.")
````

````python
tienda = TiendaOnline([], [], 0.0)

# Añadir productos al inventario
print(tienda.add_product("Guitarra eléctrica", 1200.0, 5))
print(tienda.add_product("Piano digital", 650.0, 3))
print(tienda.add_product("Violín", 180.0, 8))

print("\n--- Inventario actual ---")
print(tienda.check_inventory())

# Buscar un producto
print("\n--- Buscando producto 'Violín' ---")
tienda.search_product("Violín")

# Actualizar el stock de un producto existente
print("\n--- Actualizando stock de 'Violín' ---")
print(tienda.update_stock("Violín", 2))

# Eliminar un producto
print("\n--- Eliminando 'Piano digital' del inventario ---")
print(tienda.delete_product("Piano digital"))

# Mostrar el inventario actualizado
print("\n--- Inventario tras eliminación ---")
print(tienda.check_inventory())

# Ver el valor total del inventario
print("\n--- Valor total del inventario ---")
print(f"{tienda.inventory_value():.2f} €")

````

````python
# Realizar una compra
print("\n--- Realizar una compra ---")
tienda.purchase()
````
